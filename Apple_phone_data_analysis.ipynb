```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
```

```python
df=pd.read_csv('apple.csv')
```

```python
df
```

```python
df.columns
```

```python
df["Star Rating"].max()
```

```python
# 1 Top 5 star rating apple product
df.nlargest(5,"Star Rating")
```

```python
# Lowest 5 star rating apple phones
df.nsmallest(5,"Star Rating")
```

```python
# Ratings provided to top 5 star rating apple phones
df.nlargest(5,"Star Rating")["Star Rating"]
```

```python
# Ratings provided to lowest 5 star rating apple phones
df.nsmallest(5,"Star Rating")["Star Rating"]
```

```python
# Apple phone have highest number of reviews
df["Number Of Reviews"].max()
```

```python
# Apple phone have lowest number of reviews
df["Number Of Reviews"].min()
```

```python
# Relationship between sale price and number of ratings
style.use("ggplot")
plt.plot(df["Sale Price"],df["Number Of Ratings"],color='orange',marker='.',linestyle='--')
plt.xlabel("Sale Price",color='r')
plt.ylabel("Number Of Ratings",color='r')
plt.title("Relationship between sale price and number of ratings",color='m')
plt.legend()
```

```python
# Relationship between sale price and number of ratings
style.use("dark_background")
plt.hist([df['Sale Price'],df['Number Of Ratings']],rwidth=0.8,histtype='bar',color=['r','m'],label=['Sale Price','Number Of Ratings'])
plt.legend()
```

```python
# Relationship between sale price and number of reviews
plt.plot(df["Sale Price"],df["Number Of Reviews"],color='r',marker='.',linestyle='--')
plt.xlabel("Sale Price")
plt.ylabel("Number Of Reviews")
plt.title("Relationship between sale price and number of reviews",color='m')
plt.legend()
```

```python
# Relationship between sale price and number of reviews
plt.hist([df['Sale Price'],df['Number Of Reviews']],rwidth=0.8,histtype='bar',color=['r','m'],label=['Sale Price','Number Of Reviews'])
plt.legend()
```

```python
# Relationship between discount percentage and number of ratings
plt.plot(df["Discount Percentage"],df["Number Of Ratings"],color='r',marker='.',linestyle='--')
plt.xlabel("Discount Percentage")
plt.ylabel("Number Of Ratings")
plt.title("Relationship between discount percentage and number of ratings",color='m')
plt.legend()
```

```python
# Relationship between discount percentage and number of reviews
plt.plot(df["Discount Percentage"],df["Number Of Reviews"],color='r',marker='.',linestyle='--')
plt.xlabel("Discount Percentage")
plt.ylabel("Number Of Reviews")
plt.title("Relationship between discount percentage and number of reviews",color='m')
```
